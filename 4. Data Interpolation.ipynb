## Data Interpolation
Data interpolation is the process of estimating unknown values within a dataset based on the known values. In Python, there are various libraries available that can be used for data interpolation, such as NumPy, SciPy, and Pandas. This technique is used when we have less bumber of records in the dataset. Here is an example of how to perform data interpolation using the NumPy library:

1. Linear Interpolation
2. Cubic Interpolation With Scipy
3. Polynomial Interpolation

## Linear Interpolation

```python
import numpy as np

x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])
```

```python
import seaborn as sns
                                                                                        
sns.scatterplot(x=x,y=y)
```

```python
# we can plot using matplotlib as well

import matplotlib.pyplot as plt

plt.scatter(x,y)
```

```python
# As from the scatterplot we can understand that there is a linear relationship between x and y features
```

#### As we can see in the above graph, we have only 5 data points. If you want to increase the datapoints data points then we use "Linear Interpolation" as the initial data following the linear relationship.

```python
# Interpolate the data using the linear interpolation

x_new=np.linspace(min(x),max(x),10) # Generating the new 10 values between 1 and 5 

print(x_new)
```

```python
np.interp(x_new,x,y)
```

```python
y_interp= np.interp(x_new,x,y)
```

```python
plt.scatter(x_new,y_interp)
```

```python
# As you can observe the new data points are added and the new data points are also maintaning the linear relationship.
```

## Cubic Interpolation With Scipy

```python
# Here the relation between the two features will be in cubic in nature
```

```python
x=np.array([1,2,3,4,5])
y=np.array([1,8,27,64,125]) 
```

```python
plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d # Creating interpolation for 1D 
```

```python
# Implementing cubic interpolaion function
f=interp1d(x,y,kind="cubic")

# The string has to be one of 'linear', 'nearest', 'nearest-up', 'zero',
#     'slinear', 'quadratic', 'cubic', 'previous', or 'next'. 'zero',
#     'slinear', 'quadratic' and 'cubic' refer to a spline interpolation of
#     zeroth, first, second or third order; 'previous' and 'next' simply
#     return the previous or next value of the point; 'nearest-up' and
#     'nearest' differ when interpolating half-integers (e.g. 0.5, 1.5)
#     in that 'nearest-up' rounds up and 'nearest' rounds down. Default
#     is 'linear'.
```

```python
f
```

```python
# interpolate the data

x_new=np.linspace(min(x),max(x),10)
y_interp=f(x_new)
```

```python
y_interp
```

```python
plt.scatter(x_new,y_interp)
```

## Polynomial Interpolation

```python
x=np.array([1,2,3,4,5])
y=np.array([1,4,9,16,25])
```

```python
plt.scatter(x,y)
```

```python
# Interpolate the data using polynomial interpolation

p=np.polyfit(x,y,2) # degree of polynomial=2
```

```python
x_new=np.linspace(min(x),max(x),10) # Creating new "x" values
y_interp=np.polyval(p,x_new) # Interpolating the y values
```

```python
plt.scatter(x_new,y_interp)
```
