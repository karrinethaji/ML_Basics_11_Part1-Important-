## Standardization

### z=(Xi-mean)/standard deviation



```python
import numpy as np

mean= np.mean(df["total_bill"])
std= np.std(df["total_bill"])
```

```python
print(mean,std)
```

```python
list(df["total_bill"])
```

```python
normalized_data=[]
for i in list(df["total_bill"]):
    z_score=(i-mean)/std
    normalized_data.append(z_score)
```

```python
normalized_data
```

```python
sns.histplot(df["total_bill"],kde=True)
```

```python
sns.histplot(normalized_data,kde=True)
```

# As you can observe the shape of the data remains same for both graphs but the values got scaled for 2nd graph. For the normalized data grpah, the mean=0 and std=1 (it follows standard normal distribution)

### we can implement the standardization using the scikit learn library

https://scikit-learn.org/stable/

https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler= StandardScaler()
```

```python
scaler.fit(df[["total_bill"]]) # [[]] is mandatory

# In this fit stage, the sclaer will fit the data and will find the pre-requiste details to find the z-score i,e mean and std.
```

```python
scaler.transform(df[["total_bill"]]) # [[]] is mandatory

# transform will calculate the z-score for each observation by using the pre-requiste details which was calculated by 
# scaler object earlier
```

```python
# you can also write single line of code to replace the below two lines of code

# scaler.fit(df[["total_bill"]])
# scaler.transform(df[["total_bill"]])

# we can fit and transform the data using one step i,e

scaler.fit_transform(df[["total_bill"]])
```

```python
# Converting the standardized data into dataframe

import pandas as pd

pd.DataFrame(scaler.fit_transform(df[["total_bill"]]),columns=["total_bill_standardized"])
```

```python
df1=pd.DataFrame(scaler.fit_transform(df[["total_bill"]]),columns=["total_bill_standardized"])
```

```python
# concatenating the both the dataframes

pd.concat([df,df1],axis=1)
```

```python
# If in case we got a new test data in future, then can we transform that data using the initial scale? yes, we can using object name

# After deployment of the ML model, we use the initail object (scaler) to transform the new data.

scaler.transform([[20]]) # 20 sclaed down to the 0.0240943
```

```python
scaler.transform([[200]])
```

```python
## we can apply the standardization to multiple rows at a time.

scaler.fit_transform(df[["total_bill","tip"]])
```

```python
pd.DataFrame(scaler.fit_transform(df[["total_bill","tip"]]),columns=["total_bill_standard","tip_standardized"])
```

```python
# If in case we got a new test data in future, then can we transform that data using the initial scale? yes, we can using object name

scaler.transform([[15,6]])
```

## Normalization

We will scale down the data using the Min Max scaler technique

Min-Max= (Xi- Xmin)/(Xmax-Xmin)

The data will be scaled within [0,1]



```python
df
```

```python
# Normalization (traditional way)

min= np.min(df["total_bill"])
max=np.max(df["total_bill"])

normalized_values=[]
for i in list(df["total_bill"]):
    normalized= (i-min)/(max-min)
    normalized_values.append(normalized)
```

```python
normalized_values
```

```python
sns.histplot(df["total_bill"],kde=True)
```

```python
sns.histplot(normalized_values,kde=True)
```

```python
# As you can see the values are scaled between the 0 to 1.
```

```python
# Normalization using library

from sklearn.preprocessing import MinMaxScaler

min_max=MinMaxScaler()
```

```python
min_max.fit_transform(df[["total_bill"]])
```

```python
# we can sclae the multiple columns at a time as well

min_max.fit_transform(df[["total_bill","tip"]])
```

```python
# we can convert above columns into dataframe and we can concatenate the values with original dataframe.
```

```python
# we can also fit and transform the new data using the object name

min_max.transform([[22,4]])
```

```python
# we can also fit and transform the desired set of features

min_max.fit(df[["total_bill"]])

```

```python
min_max.transform([[23]])
```

```python
min_max.transform([[16.99]])
```
