## Missing Values
Missing values occurs in dataset when some of the informations is not stored for a variable

**Example:** If you conducted a survey for election and you are collecting the data from people to whom they have voted. But some people don't share the details like to whom they have voted. If anyone hides their opinion then in the survey we will leave that as a blank. Blank value is considered as a missing values

There are 3 mechanisms

### 1. Missing Completely at Random, MCAR:

Missing completely at random (MCAR) is a type of missing data mechanism in which the probability of a value being missing is unrelated to both the observed data and the missing data. In other words, if the data is MCAR, the missing values are randomly distributed throughout the dataset, and there is no systematic reason for why they are missing.

**For example**, in a survey about the prevalence of a certain disease, the missing data might be MCAR if the survey participants with missing values for certain questions were selected randomly and their missing responses are not related to their disease status or any other variables measured in the survey.


### 2. Missing at Random MAR:
Missing at Random (MAR) is a type of missing data mechanism in which the probability of a value being missing depends only on the observed data, but not on the missing data itself. In other words, if the data is MAR, the missing values are systematically related to the observed data, but not to the missing data.
Here are a few examples of missing at random:

**Income data:** Suppose you are collecting income data from a group of people, but some participants choose not to report their income. If the decision to report or not report income is related to the participant's age or gender, but not to their income level, then the data is missing at random. Some people don't want to repprt their gender details during the survey.

**Medical data:** Suppose you are collecting medical data on patients, including their blood pressure, but some patients do not report their blood pressure. If the patients who do not report their blood pressure are more likely to be younger or have healthier lifestyles, but the missingness is not related to their actual blood pressure values, then the data is missing at random.

## 3. Missing data not at random (MNAR)
It is a type of missing data mechanism where the probability of missing values depends on the value of the missing data itself. In other words, if the data is MNAR, the missingness is not random and is dependent on unobserved or unmeasured factors that are associated with the missing values.

**For example**, suppose you are collecting data on the income and job satisfaction of employees in a company. If employees who are less satisfied with their jobs are more likely to refuse to report their income, then the data is not missing at random. In this case, the missingness is dependent on job satisfaction, which is not directly observed or measured.

## ** Imputation Missing Values ( widely used as a data scientist )

### 1. Mean value Imputation:

Mean imputation (or mean substitution) replaces missing values of a certain variable by the mean of non-missing cases of that variable.

```python
sns.distplot(df["age"])
```

```python
sns.histplot(df["age"],kde=True)
```

As we can see age column following almost normal distribution and age column has 177 missing values as well. So, we can fill the nan values with the mean of the age column data.

### NOTE:

##### Mean Imputation works well when we have a normal distributed data.

```python
df["age"].fillna(df["age"].mean())
```

```python
df["age_impute_by_mean"]=df["age"].fillna(df["age"].mean())
```

```python
df[["age","age_impute_by_mean"]]
```

```python
df[["age","age_impute_by_mean"]].isnull().sum()
```

#### As you can see, the nan values in the column "age" are replaced by the mean of the age column

#### Summary:
    
Mean Imputation works well when we have a normal distributed data.

If the distribution follows non-normal distribution like left skewed or right skewed distribution then obviously we will have outliers in the columns. so we will use the median imputation.

### 2. Median Value Imputation (If we have outliers in the dataset)

Median imputation replaces missing values of a certain variable by the median of non-missing cases of that variable.

```python
df["age_impute_by_median"]= df["age"].fillna(df["age"].median())
```

```python
df[["age","age_impute_by_median"]]
```

### 3. Mode Value Imputation (applied on categorical feature)

```python
df[df["embarked"].isnull()]
```

"embarked" is an example for Missing Completely at Random, MCAR

```python
# finding unique values

df["embarked"].unique()
```

**Here we can replace the nan value with the highest repeated element in the "embarked" column**

```python
# Finding non nan values

df["embarked"].notna()
```

```python
df[df["embarked"].notna()]
```

```python
# From the non nan records extract the embarked column and find the mode.

df[df["embarked"].notna()]["embarked"]
```

```python
df[df["embarked"].notna()]["embarked"].mode()
```

```python
df[df["embarked"].notna()]["embarked"].mode()[0]
```

```python
mode_value= df[df["embarked"].notna()]["embarked"].mode()[0]
```

```python
# filling the mode value in place of nan value in the "embarked" column.

df["embarked_mode_imputation"]=df["embarked"].fillna(mode_value)
```

```python
df[["embarked","embarked_mode_imputation"]]
```

```python
df[["embarked","embarked_mode_imputation"]].isnull().sum()
```

```python
# As you can see there are no nan values under "embarked_mode_imputation" column.
```
