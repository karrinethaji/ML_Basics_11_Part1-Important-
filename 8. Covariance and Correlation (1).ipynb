```python
# Let's use different dataset healthexp to illuistrate the covariance and correlation concept

sns.load_dataset("healthexp")
```

```python
df=sns.load_dataset("healthexp")
```

```python
df
```

```python
sns.scatterplot(x="Spending_USD",y="Life_Expectancy",data=df)

# From the scatter plot we can understand the Spending_USD and Life_Expectancy has positive correlation. as Spending_USD increases,
# Life_Expectancy also increasing.
```

```python
# Finding covaraince using numpy

import numpy as np

np.cov([df.Spending_USD,df.Life_Expectancy]) # Output is in the form of array

# we can also find the covariance using pandas
```

```python
df[["Year","Spending_USD","Life_Expectancy"]].cov()
```

```python
# correlation

#Pearsons correlation

import seaborn as sns

df[["Year","Spending_USD","Life_Expectancy"]].corr() 

# by default the mrthod="pearson"
```

```python
df[["Year","Spending_USD","Life_Expectancy"]].corr(method="pearson") 
```

```python
#Spearmen rank correlation

df[["Year","Spending_USD","Life_Expectancy"]].corr(method="spearman") 
```

```python
# As you can see there is positive correlation between the Year and Life_Expectancy

sns.scatterplot(x="Year",y="Life_Expectancy",data=df)
```

```python
# As you can see there is positive correlation between the Spending_USD and Life_Expectancy

sns.scatterplot(x="Spending_USD",y="Life_Expectancy",data=df)
```

```python
# correlation=1 between the same features like Life_Expectancy vs Life_Expectancy

sns.scatterplot(x="Life_Expectancy",y="Life_Expectancy",data=df)
```

## Observations:

Year feature has **positive correlation** with Life_Expectancy (0.896), i,e **as year increases Life_Expectancy also increases.**

Spending_USD feature has also **positive correlation** with Life_Expectancy (0.747) 

Correlation between the same features will always be 1 i,e Year vs Year, Life_Expectancy vs Life_Expectancy
